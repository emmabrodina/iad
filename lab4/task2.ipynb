**Task 2. Hugging Face / Pipeline module / Ukrainian models**

```python
from transformers import pipeline
```

```python
# Переклад тексту (Translation)
translator = pipeline("translation_en_to_uk", model="Helsinki-NLP/opus-mt-en-uk")
translated_text = translator("Hello, how are you?")[0]['translation_text']
print("Переклад тексту:", translated_text)
```

```python
# Аналіз настроїв (Sentiment Analysis)
classifier = pipeline("sentiment-analysis", model="cardiffnlp/twitter-xlm-roberta-base-sentiment")
result = classifier("Цей продукт мені дуже сподобався!")[0]
print(f"Аналіз настроїв: {result['label']} з ймовірністю {result['score']:.2f}")
```

```python
# Розпізнавання сутностей NER
ner_pipeline = pipeline("ner", model="xlm-roberta-large-finetuned-conll03-english", aggregation_strategy="simple")

# Тестовий текст
text = "Володимир Зеленський є президентом України."
ner_results = ner_pipeline(text)

# Вивід результатів
print("Розпізнавання іменованих сутностей:")
for res in ner_results:
    print(f"Слово: {res['word']}, Мітка: {res['entity_group']}, Ймовірність: {res['score']:.2f}")
```

```python
# Питання-відповідь (QA)
qa_pipeline = pipeline("question-answering", model="deepset/xlm-roberta-base-squad2")
context = "Тарас Шевченко був відомим українським поетом, письменником та художником."
question = "Хто був Тарас Шевченко?"
qa_result = qa_pipeline(question=question, context=context)
print(f"Відповідь на запитання: {qa_result['answer']}")
```
